# Visualize

* Learn to use Matplotlib
* Subplots
* Multiple plots
* Bar plots

```python
import pandas as pd
```

```python
data = pd.read_csv("AAPL.csv", index_col = 0, parse_dates = True)
```

```python
data.head()
```

```python
data.plot()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

%matplotlib inline is suitable for static plots that are generated once and do not need to be modified interactively, while %matplotlib notebook is more suitable for interactive plots that require zooming, panning, and other interactive features.

```python
data.plot()
```

```python
data['Close'].plot()
```

```python
fig, ax = plt.subplots()
data['Close'].plot(ax=ax)
ax.set_ylabel('Price')
ax.set_title("AAPL")
```

```python
fig, ax = plt.subplots(2,2)
data['Open'].plot(ax=ax[0,0], title='Open')
data['High'].plot(ax=ax[0,1], title='High')
data['Low'].plot(ax=ax[1,0], title='Low')
data['Close'].plot(ax=ax[1,1], title = 'Close')
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
data['Volume'].loc['2020-07-01':'2020-08-15'].plot.barh(ax=ax)
```
